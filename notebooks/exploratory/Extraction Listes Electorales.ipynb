```python
import requests 
from bs4 import BeautifulSoup
import pandas as pd
import datetime as datetime
import time
```

```python
def get_json(url):
    lu = 'non'
    while lu == 'non' :
        try :
            r = requests.get(url)
            lu = 'oui'
        except :
            print('On réessaie')
    r = r.text
    d = pd.read_json(r)
    return d
```

```python
url = 'http://www.ceni-niger.org/ceniProject/regions/reg'
regions = get_json(url)
```

```python
url = "http://www.ceni-niger.org/ceniProject/departements/get/1"
departements = get_json(url)
```

```python
for i in range(1 , len(regions)) :
    print(regions.NOM_REGION[i])
    url = "http://www.ceni-niger.org/ceniProject/departements/get/" + str(regions.ID_REGION[i])
    dep = get_json(url)
    departements = departements.append(dep, ignore_index=True)
```

```python
url = "http://www.ceni-niger.org/ceniProject/communes/get/101"
communes = get_json(url)
```

```python
for i in range(1 , len(departements)) :
    print(departements.NOM_DEPART[i])
    url = "http://www.ceni-niger.org/ceniProject/communes/get/" + str(departements.ID_DEPART[i])
    dep = get_json(url)
    communes = communes.append(dep, ignore_index=True)
```

```python
url = "http://www.ceni-niger.org/ceniProject/bureaux/get/10101"
bureaux = get_json(url)

for i in range(1 , len(communes)) :
    print(communes.NOM_COMMUNE[i])
    url = "http://www.ceni-niger.org/ceniProject/bureaux/get/" + str(communes.ID_COMMUNE[i])
    dep = get_json(url)
    bureaux = bureaux.append(dep, ignore_index=True)
```
