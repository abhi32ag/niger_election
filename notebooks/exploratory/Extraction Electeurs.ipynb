```python
import requests 
from bs4 import BeautifulSoup
import pandas as pd
import datetime as datetime
import time
```

```python
regions = pd.read_csv('data/Niger_Regions.csv' , encoding = "ISO-8859-1")
departements = pd.read_csv('data/Niger_Departements.csv' , encoding = "ISO-8859-1")
communes = pd.read_csv('data/Niger_Communes.csv' , encoding = "ISO-8859-1")
bureaux = pd.read_csv('data/Niger_Bureaux.csv' , encoding = "ISO-8859-1")

full_dat = pd.merge(regions , departements)
full_dat = pd.merge(full_dat , communes)
full_dat = pd.merge(full_dat , bureaux)
```

```python
def extract_list(url) :
    lu  = 'non'
    while lu == 'non' :
        try :
            r = requests.get(url)
            lu = 'oui'
        except :
            print('On r√©essaye')
    r = r.text
    soup = BeautifulSoup(r , 'html.parser') 
    table_results =  soup.select("table") 
    table_rows = table_results[0].findAll('tr')
    nrows = len(table_rows)
    
    dframe = {}
    
    for row in range(1 , nrows) :
        cont = table_rows[row].findAll('td')
        row_val = []
        ncol = len(cont)
        for column in range(ncol) :
            row_val = row_val + [cont[column].text.strip()]

        dframe[str(row)] = row_val

    out = pd.DataFrame(dframe).T
    out 
    return out
```

```python
n_communes = len(communes)
for i in range(0,n_communes) :
    nom_commune = communes.NOM_COMMUNE[i]
    id_commune = communes.ID_COMMUNE[i]
    dat_com = bureaux[bureaux.ID_COMMUNE == id_commune]
    n_bureaux = len(dat_com)
    n_bureaux_extracted = 0
    print(nom_commune)
    for j in range(n_bureaux):
        id_bureau = dat_com.reset_index().ID_BUREAU[j]
        url = 'http://ceni-niger.org/ceniProject/electeurs/' + str(id_bureau) + '/bureau'
        print(url)
        electeurs = extract_list(url)
        if len(electeurs) > 0:
            electeurs.columns = ['0','1','2','3','4','5','6']
            n_bureaux_extracted = n_bureaux_extracted + 1
        if n_bureaux_extracted == 1 :
            electeurs['ID_BUREAU'] = str(id_bureau)
            out = electeurs
        if n_bureaux_extracted > 1 :
            electeurs['ID_BUREAU'] = str(id_bureau)
            out = out.append(electeurs, ignore_index=False)
    nom_commune = nom_commune.replace('/','-')
    nom_csv = 'data/listes_electeurs/' + nom_commune + '.csv'
    print(str(100 * (i+1) / (n_communes)) + ' % extracted')
    out.to_csv(nom_csv)
```
