```python
def extract_region(url) :
    lu  = 'non'
    while lu == 'non' :
        try :
            r = requests.get(url)
            lu = 'oui'
        except :
            print('On réessaye')
    r = r.text
    soup = BeautifulSoup(r , 'html.parser') 
    table_results =  soup.select("table") 
    table_rows = table_results[0].findAll('tr')
    nrows = len(table_rows)
    
    dframe = {}
    reg = soup.select("h1")[0].text.strip()
    print('Extracting ' + reg)
    
    for row in range(1 , nrows) :
        cont = table_rows[row].findAll('td')
        row_val = [reg]
        ncol = len(cont)
        for column in range(ncol) :
            row_val = row_val + [cont[column].text.strip()]

        dframe[str(row)] = row_val
    
    cols = ['ID']
    colnames = table_rows[0].findAll('th')
    ncol = len(colnames)
    for column in range(ncol) :
        cols = cols + [colnames[column].text.strip()]
    
    out = pd.DataFrame(dframe).T
    if len(out) > 0 :
        out.columns = cols
    out['time'] = datetime.datetime.now()
    return out
```

```python
def extract_all(url_base , ids) :
    data_out = []
    for i in ids :
        print(i)
        url = url_base + str(i)
        out = extract_region(url)
        out['zone_id'] = str(i)
        if len(data_out) > 0 :
            if len(out) > 0 :
                data_out = data_out.append(out, ignore_index=True)
        if len(data_out) == 0 :
            data_out = out

    return data_out        
```

```python
def make_num(u) :
    num_data = ['Inscrits', 'Inscrits ayant voté', 'Votants sur liste additive', 'Nbre total de votants', 
                'Bulletins blancs ou nuls', 'Suffrages exprimés valables' , 'Laouan Magagi', 'Mahaman Jean Philipe Padonou', 
                'Abdou Labo', 'Kassoum M. Moctar', 'Adal Rhoubeid', 'Mahamane Ousmane', 'Seyni Omar', 'Tahirou Guimba', 
                'Hama Amadou', 'Ibrahim Yacouba',  'Mahaman Hamissou Maman','Mahamadou Issoufou', 'Dr. Abdoulaye Amadou Traoré', 
                'Cheffou Amadou','Boubacar Cissé']
    for col in num_data :
        u[col] = u[col].str.replace(' ', '').astype(int)
    return u
```
